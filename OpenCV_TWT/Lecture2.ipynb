```python
from google.colab.patches import cv2_imshow
import cv2

img = cv2.imread('/content/jagrit.jpeg');

print(img)#pixels of img are loaded into numpy array)
```

```python
print(type(img))
```

```python
#open cv uses bgr format (blue,green,red)
# [0,0,0] = white and [255,255,255] = black
print(img[100][40])

# adding random pixels to a pic
import random as r
for i in range(100):
  for j in range(img.shape[1]):#represents the width of image ,,,,if 0 was there it represents height of image
    img[i][j] = [r.randint(0,255),r.randint(0,255),r.randint(0,255)];
cv2_imshow(img);
cv2.waitKey(0);
cv2.destroyAllWindows()
```

```python
#copy and pasting from img
tag = img[250:350 , 310:450];
img[200:300 , 300:440] = tag;
cv2_imshow(img);
cv2_imshow(tag);
```
